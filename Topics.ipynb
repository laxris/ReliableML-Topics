```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

# manuelly transfered top elements and aggregated
objects = ('real time', 'large scale', 'cellular network', 'data center', 'wireless sensor', 'QoS',
           'virtual network', 'SDN', 'base station', 
           'centric networking', "machine learning", "ILP", "could computing", "convex optim")

counts = [67, 60, 49, 44, 42, 42, 40, 38, 37, 35, 34, 32, 30, 15]

# topics = black, methods = gray
colors=['black', 'black', 'black', 'black', 'black', 'black', 'black', 
                                          'black', 'black', 'black', 'grey', 'grey', 'black', 'grey',]

x_pos = np.arange(len(objects))

width = 1/1.5
plt.bar(x_pos, counts, width, color=colors)
plt.xticks(x_pos + 0.25, objects, rotation='vertical')
plt.ylabel('count')
plt.tight_layout()

#plt.savefig('bar.png')
plt.show()
```
